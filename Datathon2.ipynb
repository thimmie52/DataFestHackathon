```python
threshold = np.mean(df['Transaction Amount'])
threshold_range = np.sqrt(np.var(df['Transaction Amount']))

df['Transaction Levels'] = df['Transaction Amount'].apply(lambda x: 'Above medium' if x > threshold + threshold_range else('Below med' if  x < threshold - threshold_range else 'medium'))
```

```python
threshold = np.mean(df['User Income'])
threshold_range = np.sqrt(np.var(df['User Income']))

df['Income Levels'] = df['User Income'].apply(lambda x: 'Above medium' if x > threshold + threshold_range else('Below med' if  x < threshold - threshold_range else 'medium'))
```

```python
df['TransactionPerSalary '] = df['Transaction Amount'] / df['User Income']
```

```python
df['Transaction Date and Time'] = pd.to_datetime(df['Transaction Date and Time'])
df = df.sort_values(by = ['User ID', 'Transaction Date and Time'])
```

```python
current_value = None
current_index = -1
indices = []

# Iterate through the rows of the column
for value in df['User ID']:
    if value == current_value:
        current_index += 1
    else:
        current_value = value
        current_index = 0
    indices.append(current_index)

# Add the indices as a new column to the DataFrame
df['No_of_transfers_before'] = indices
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.2,
    random_state=24,
    stratify = y,
    shuffle =True
)
```

```python
x_columns = []
numerical_columns = []
categorical_columns = []
for col in x.columns:
  if str(x[col].dtype).split("'")[0] == 'float64' or str(x[col].dtype).split("'")[0] == 'int64':
    numerical_columns.append(col)
  else:
    categorical_columns.append(col)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}


for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    x_train[column] = label_encoders[column].fit_transform(x_train[column])

for column in categorical_columns:
    x_test[column] = label_encoders[column].transform(x_test[column])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
encoders = {}

# Fit and transform x_train
for column in numerical_columns:
    encoders[column] = StandardScaler()
    x_train[column] = encoders[column].fit_transform(x_train[column].values.reshape(-1, 1))

# Transform x_test using the same encoders
for column in numerical_columns:
    x_test[column] = encoders[column].transform(x_test[column].values.reshape(-1, 1))
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier



# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

# Define the number of folds (k)
k = 5

# Create a KFold object for splitting data
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Perform 5-fold cross-validation
scores = cross_val_score(clf, x_train, y_train, cv=kf)

# Print the cross-validation scores
print("Cross-Validation Scores:", scores)

# Calculate and print the average score
avg_score = sum(scores) / k
print("Average Score:", avg_score)
```
